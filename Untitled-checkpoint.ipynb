```python
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})  # Codificar género
print(df.isnull().sum())  # Verificar valores nulos = df.drop(columns=["
df.fillna(df.mean(), inplace=True)  # Reemplazar valores nulos por la media
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']] = scaler.fit_transform(df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])
```

```python
from sklearn.feature_selection import mutual_info_regression

features = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
mi = mutual_info_regression(df[features], df['Spending Score (1-100)'])
feature_importance = pd.Series(mi, index=features).sort_values(ascending=False)
print(feature_importance)
```

```python


# Crear matriz de enlace
linkage_matrix = linkage(df[['Annual Income (k$)', 'Spending Score (1-100)']], method='ward')

# Asignar clústeres
df['Cluster'] = fcluster(linkage_matrix, t=4, criterion='maxclust')
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(df[['Annual Income (k$)', 'Spending Score (1-100)']], df['Cluster'])
print("Coeficiente de Silhouette:", silhouette_avg)
```

```python
from sklearn.metrics import calinski_harabasz_score

ch_score = calinski_harabasz_score(df[['Annual Income (k$)', 'Spending Score (1-100)']], df['Cluster'])
print("Índice de Calinski-Harabasz:", ch_score)
```

```python
from scipy.cluster.hierarchy import dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linkage_matrix)
plt.title("Dendrograma de Clustering Jerárquico")
plt.show()
```

```python
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', data=df, palette='viridis')
plt.title("Distribución de los Clústeres")
plt.show()
```
